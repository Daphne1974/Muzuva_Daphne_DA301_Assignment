```python
# To create the OLS (Ordinary Least Squares) multiple linear regression model, 
# you can use the statsmodels library in Python.
import statsmodels.api as sm

# Define the independent variables (predictors)
X = imported_reviews[["age", "remuneration (k£)", "spending_score (1-100)"]]

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Define the dependent variable (target)
y = imported_reviews["loyalty_points"]

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Get the summary of the regression model
print(model.summary())

# In this code:
# We first define the independent variables (predictors) X as a DataFrame containing the columns "age," "remuneration (k£)," and "spending_score (1-100)."
# We add a constant term to the model using sm.add_constant. This constant represents the intercept of the linear regression equation.
# We define the dependent variable (target) y as the "loyalty_points" column.
# We use sm.OLS to fit the multiple linear regression model, and fit() to actually estimate the model parameters.
# Finally, we print the summary of the regression model using model.summary(), which provides detailed statistics 
# and information about the model, including coefficients, p-values, R-squared, and more.
# The model object now contains your multiple linear regression model, and you can use it to 
# make predictions or further analyze the relationship between the independent variables and the dependent variable.
```

```python
# To determine the optimal number of clusters for k-means clustering, Using the Silhouette method and the Elbow method. 
# Create the plots with titles and legends:

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Create a range of potential cluster numbers to explore
cluster_range = range(2, 11)  
# Try 2 to 10 clusters

# Initialize lists to store silhouette scores and inertias
silhouette_scores = []
inertia_values = []

# Calculate silhouette scores and inertias for each cluster number
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(df2)
    silhouette_avg = silhouette_score(df2, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    inertia_values.append(kmeans.inertia_)

# Plot the Silhouette method results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(cluster_range, silhouette_scores, marker='o', linestyle='-', color='b')
plt.title("Silhouette Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)

# Plot the Elbow method results
plt.subplot(1, 2, 2)
plt.plot(cluster_range, inertia_values, marker='o', linestyle='-', color='r')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.grid(True)

# Show the combined plot
plt.tight_layout()
plt.show()

# In this code:
# We specify a range of potential cluster numbers to explore, from 2 to 10 clusters.
# We calculate the Silhouette scores and inertias for each cluster number by fitting k-means clustering models to the data.
# We create two subplots, one for the Silhouette method results and one for the Elbow method results.
# In the Silhouette subplot, we plot the Silhouette scores for different cluster numbers. 
# The optimal number of clusters is typically associated with the highest Silhouette score.
# In the Elbow subplot, we plot the inertias for different cluster numbers. 
# The "elbow" point, where the inertia begins to level off, indicates a good choice for the number of clusters.
# To determine the optimal number of clusters:
# Silhouette Method: Look for the number of clusters that gives the highest Silhouette score. 
# A higher Silhouette score indicates better-defined clusters.
# Elbow Method: Look for the "elbow" point in the plot of inertia values. 
# The "elbow" is where the inertia begins to level off, suggesting that additional clusters may not significantly 
# reduce the variance within clusters.
```

```python
# To evaluate the usefulness of at least three values for k based on insights from the Elbow and Silhouette methods, 
# and to plot the predicted k-means, 
# Elbow Method and Silhouette Method Insights: Already calculated the Silhouette scores and inertias for a range 
# of cluster numbers. Now, we can visually inspect the plots to identify potential values for k. 
# In general, we look for points of inflection or significant changes in the plots.
# Plot Predicted K-Means: After determining candidate values for k, fit k-means clustering models to the data using 
# these values and visualize the predicted clusters.

# Based on insights from the Elbow and Silhouette methods, consider at least three values for k.
k_values = [3, 4, 5]  
# We can adjust these values based on the insights.

# Initialize a list to store cluster assignments for each k
cluster_assignments = []

# Plot the predicted k-means for each value of k
plt.figure(figsize=(12, 4))

for i, k in enumerate(k_values, 1):
    # Fit a k-means model with the current value of k
    kmeans = KMeans(n_clusters=k, random_state=0)
    cluster_labels = kmeans.fit_predict(df2)
    cluster_assignments.append(cluster_labels)
    
    # Create a subplot for each k
    plt.subplot(1, len(k_values), i)
    
    # Plot the predicted clusters
    plt.scatter(df2["remuneration (k£)"], df2["spending_score (1-100)"], c=cluster_labels, cmap='viridis')
    plt.title(f"K-Means Clustering (k={k})")
    plt.xlabel("Remuneration (k£)")
    plt.ylabel("Spending Score (1-100)")
    
plt.tight_layout()
plt.show()

# In this code:
# We select three values for k (3, 4, and 5) based on insights from the Elbow and Silhouette methods 
# we can adjust these values as needed.
# We loop through each value of k, fit a k-means model to the data, and store the cluster assignments 
# in the cluster_assignments list.
# We create subplots for each value of k and plot the predicted clusters using different colors.
# To determine which value of k might give you the best clustering, consider the following:
# Elbow Method: Look at the Elbow plot.  A clear "elbow" point, indicates that k value could be a good choice. 
# It's where the inertia begins to level off.
# Silhouette Method: Look at the Silhouette scores for each k. A higher Silhouette score indicates better-defined 
# clusters. Choose the k with the highest Silhouette score.
# Visual Inspection: Examine the plots of the predicted clusters. Consider how well the clusters make sense in the 
# context of our data and objectives. A good k should result in meaningful and well-separated clusters.
# Ultimately, the best choice of k will depend on the specific characteristics of your data and your clustering goals. 
# We may need to balance simplicity (fewer clusters) with meaningfulness (clusters that capture distinct patterns).
```

```python
# To determine the frequency distribution and polarity of words in the text data:
# Tokenize the text data.
# Create a frequency distribution of words.
# Remove alphanumeric characters and stopwords.
# Create a word cloud without stopwords.
# Identify the 15 most common words.
# Calculate their sentiment polarity using a sentiment analysis tool.
# you can use column names from the DataFrame for text data analysis. 
# DataFrame with columns named 'review' or 'summary,' you can extract and analyze text data from those columns. 

import nltk
import pandas as pd
from nltk import FreqDist
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.sentiment import SentimentIntensityAnalyzer

# Load your data from a CSV file into a DataFrame
df = pd.read_csv('turtle_reviews.csv')  

# Choose the column you want to analyze ('review' or 'summary')
text_column = 'review'  # Change to 'summary' to analyze the 'summary' column

# Extract the text data from the chosen column
text_data = df[text_column].str.cat(sep=' ')

# Tokenize the text
tokens = nltk.word_tokenize(text_data)

# Create a frequency distribution of words
freq_dist = FreqDist(tokens)

# Remove alphanumeric characters and stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]

# Create a word cloud without stopwords
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered_words))

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title(f"Word Cloud for {text_column.capitalize()} Without Stopwords")
plt.show()

# Identify the 15 most common words
common_words = freq_dist.most_common(15)
print(f"15 Most Common Words in {text_column.capitalize()}:")
for word, freq in common_words:
    print(f"{word}: {freq}")

# Calculate sentiment polarity for the 15 most common words
sia = SentimentIntensityAnalyzer()
sentiment_scores = {}

for word, _ in common_words:
    sentiment = sia.polarity_scores(word)
    sentiment_scores[word] = sentiment['compound']

# Display sentiment polarity for each of the 15 most common words
print(f"\nSentiment Polarity for the 15 Most Common Words in {text_column.capitalize()}:")
for word, polarity in sentiment_scores.items():
    print(f"{word}: {polarity}")

# In this code, you can specify whether you want to analyze the 'review' or 'summary' column by setting the text_column 
# variable accordingly. The code will then extract, analyze, and visualize the text data from the chosen column 
# This code tokenizes the text, creates a frequency distribution of words, removes alphanumeric characters and stopwords, 
# generates a word cloud, identifies the 15 most common words, and calculates their sentiment polarity using the 
# SentimentIntensityAnalyzer from NLTK.
```
